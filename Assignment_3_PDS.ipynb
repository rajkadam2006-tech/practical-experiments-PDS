1. What is Data Wrangling? Explain the difference between Data Wrangling and Data Preprocessing.

Data wrangling is the process of cleaning, transforming, and organizing raw data into a structured and usable format for analysis. In real-world situations, data is often incomplete, inconsistent, or contains errors, so data wrangling helps improve its quality by handling missing values, removing duplicates, and converting data into the required format. Data wrangling is different from data preprocessing in terms of purpose and stage. Data wrangling focuses on preparing raw data for analysis by cleaning and organizing it, whereas data preprocessing is mainly used to prepare data for machine learning models. Data preprocessing includes tasks such as scaling, normalization, and encoding. In simple terms, data wrangling makes data analysis-ready, while preprocessing makes it model-ready.

2. What is the importance of Exploratory Data Analysis (EDA) in Pandas?

Exploratory Data Analysis (EDA) is an important step in data analysis that helps in understanding the structure and characteristics of a dataset. It allows us to identify missing values, detect patterns and relationships, and find outliers or errors in the data. By performing EDA, we can gain insights into the dataset before applying any advanced analysis or modeling techniques. In Pandas, functions such as head(), info(), and describe() are commonly used for EDA. These functions help in viewing the data, understanding data types, and summarizing statistical information. Overall, EDA improves decision-making and ensures that the data is suitable for further analysis.

3. What are the main differences between Series, DataFrame, and Panel?

In Pandas, Series, DataFrame, and Panel are different types of data structures used to store and manipulate data. A Series is a one-dimensional array that can hold data of any type and is similar to a single column. A DataFrame is a two-dimensional structure consisting of rows and columns, making it suitable for storing tabular data and performing analysis. A Panel, on the other hand, is a three-dimensional data structure used for handling multi-dimensional data. However, Panel is now deprecated in modern Pandas and is no longer commonly used. Thus, Series is used for simple data, DataFrame is the most widely used structure for analysis, and Panel was used for higher-dimensional data.

4. Explain the use of isnull(), notnull(), dropna(), and fillna().

In Pandas, several functions are used to handle missing data. The isnull() function is used to detect missing values in a dataset and returns True for missing entries. The notnull() function is the opposite of isnull() and returns True for non-missing values. The dropna() function is used to remove rows or columns that contain missing values, which is useful when the missing data is not significant. On the other hand, the fillna() function is used to replace missing values with a specific value, such as the mean, median, or a constant. These functions are essential for cleaning data and ensuring accurate analysis.

5. What is column transformation and why is it important in data analysis?

Column transformation is the process of modifying or creating new columns in a dataset to make the data more suitable for analysis. It includes operations such as changing data types, creating new columns from existing ones, applying mathematical operations, and encoding categorical data into numerical form. Column transformation is important because it improves data quality, ensures consistency, and makes the dataset easier to analyze. It also helps in preparing data for machine learning models and enhances visualization and interpretation of results. Therefore, column transformation plays a crucial role in effective data analysis.

```python
import pandas as pd
import numpy as np
# Create Dataset (Raw data with missing values)
data = {
    'Name': ['Amit', 'Riya', 'John', 'Sara', 'Raj'],
    'Age': [20, 21, np.nan, 23, 22],
    'Marks': [85, np.nan, 78, 92, 88]
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)

#Exploratory Data Analysis (EDA)
print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

#Check missing value
print("\nMissing Values (True = Missing):")
print(df.isnull())

print("\nCount of Missing Values:")
print(df.isnull().sum())

#Use of notnull()
print("\nNon-Missing Values (True = Present):")
print(df.notnull())

#Remove Missing Values using dropna()
df_drop = df.dropna()

print("\nDataset after removing missing values:")
print(df_drop)

# Fill Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill Marks with mean
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())

print("\nDataset after filling missing values:")
print(df)

#Column Transformation
# Create new column
df['Total'] = df['Marks'] + 10

# Convert datatype of Age to integer
df['Age'] = df['Age'].astype(int)

print("\nDataset after column transformation:")
print(df)
```

### **Data Wrangling Using IRIS Dataset**

```python
import pandas as pd
import numpy as np
```

```python
# Load dataset from URL
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

df.head()
```

```python
# First 5 rows
df.head()

# Last 5 rows
df.tail()

# Dataset information
df.info()

# Statistical summary
df.describe()
```

```python
df.isnull().sum()
```

```python
# Introduce missing values manually
df.loc[0, 'sepal_length'] = np.nan
df.loc[1, 'petal_width'] = np.nan

df.isnull().sum()
```

```python
# Fill missing values with mean
df['sepal_length'].fillna(df['sepal_length'].mean(), inplace=True)
df['petal_width'].fillna(df['petal_width'].mean(), inplace=True)

df.isnull().sum()
```

```python
df.notnull().head()
```

```python
# Create new column
df['sepal_area'] = df['sepal_length'] * df['sepal_width']

df.head()
```

```python
# Filter rows where sepal_length > 5
df[df['sepal_length'] > 5]
```

```python
# Average values by species
df.groupby('species').mean()
```

```python
# Sort by sepal_length
df.sort_values(by='sepal_length', ascending=False).head()
```
