**Theory :**
1. What is web scraping? Explain the role of the Requests library in web scraping.

Web scraping is the process of extracting data from websites automatically using programs or scripts. It allows users to collect large amounts of data such as text, images, or links from web pages for analysis or other purposes. The Requests library in Python plays an important role in web scraping by sending HTTP requests to websites and retrieving their HTML content. It acts as a bridge between the user and the web server, allowing the program to access web pages just like a browser. Once the HTML content is fetched using Requests, it can be further processed and analyzed using parsing libraries like BeautifulSoup.

2. What is BeautifulSoup? How does it help in parsing HTML?

BeautifulSoup is a Python library used for parsing HTML and XML documents. It helps in extracting specific data from web pages by navigating through the HTML structure in a simple and readable way. BeautifulSoup converts complex HTML code into a tree-like structure, making it easy to locate elements such as tags, classes, and IDs. With its built-in functions, users can search, filter, and extract required data efficiently. It is especially useful when dealing with messy or poorly structured HTML, as it automatically handles errors and simplifies the parsing process.

3. Differentiate between web scraping and web crawling.

Web scraping and web crawling are related but different concepts. Web scraping refers to extracting specific data from a particular website or web page, focusing on collecting useful information like prices, text, or links. On the other hand, web crawling involves automatically browsing or indexing multiple web pages across the internet, usually by following links from one page to another. Crawlers are used by search engines to index websites, whereas scraping is typically used for data extraction from targeted sources. In simple terms, crawling is about discovering pages, while scraping is about extracting data from them.

4. What are the commonly used methods in BeautifulSoup?

BeautifulSoup provides several useful methods to extract and navigate data from HTML documents. Some commonly used methods include find() and find_all() for locating specific tags or elements, get_text() for extracting text content, and get() for accessing attribute values like links or image sources. Methods like select() allow users to use CSS selectors to target elements. Additionally, navigation methods such as .parent, .children, and .next_sibling help in moving through the HTML structure. These methods make data extraction flexible and efficient.

5. What are the ethical and legal considerations in web scraping?

Web scraping must be performed responsibly by following ethical and legal guidelines. Users should always check a website’s robots.txt file to understand which parts of the site can be accessed or scraped. It is important to avoid sending too many requests in a short time, as this can overload the website’s server. Scraping copyrighted or sensitive data without permission may lead to legal issues. Additionally, personal or private data should never be collected without consent. Respecting website terms of service and using scraped data responsibly ensures ethical web scraping practices.

Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Send Request to website

```python
url = "http://books.toscrape.com/"

response = requests.get(url)

print("Status Code:", response.status_code)
```

Parse HTML content

```python
soup = BeautifulSoup(response.text, 'html.parser')

print(soup.title.text)
```

Extract Book Titles

```python
books = soup.find_all('h3')

titles = []

for book in books:
    title = book.find('a')['title']
    titles.append(title)

titles
```

Extract Prices

```python
prices = soup.find_all('p', class_='price_color')

price_list = []

for price in prices:
    price_list.append(price.text)

price_list
```

Extract Rating

```python
ratings = soup.find_all('p', class_='star-rating')

rating_list = []

for rating in ratings:
    rating_list.append(rating['class'][1])

rating_list
```

Create DataFrame

```python
data = pd.DataFrame({
    'Title': titles,
    'Price': price_list,
    'Rating': rating_list
})

data
```

Save Data to CSV

```python
data.to_csv("books_data.csv", index=False)

print("Data saved successfully!")
```
