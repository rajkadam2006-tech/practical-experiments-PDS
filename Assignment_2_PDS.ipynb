1. What is data ingestion and why is it important in data science? Which Python library is commonly used for data ingestion and cleaning?

Data ingestion is the process of collecting and importing data from various sources such as databases, CSV files, APIs, or web services into a system for further processing and analysis. It is a fundamental step in the data science pipeline because the quality and availability of data directly impact the accuracy and effectiveness of analysis and machine learning models. Without proper data ingestion, it is not possible to work with real-world data efficiently. In Python, one of the most commonly used libraries for data ingestion and cleaning is Pandas. It provides powerful data structures like DataFrames and functions that make it easy to read, manipulate, and clean data from different formats such as CSV, Excel, and JSON files.

2. Why is data cleaning necessary before performing analysis or machine learning?

Data cleaning is necessary because real-world data is often incomplete, inconsistent, and contains errors such as missing values, duplicate records, or incorrect formats. If such unclean data is used directly for analysis or machine learning, it can lead to inaccurate results, poor model performance, and misleading conclusions. Data cleaning ensures that the dataset is reliable, consistent, and properly formatted, which improves the quality of insights and predictions. It involves tasks like handling missing values, removing duplicates, correcting errors, and standardizing data formats. Therefore, data cleaning is a critical step to ensure the success of any data science project.

3. How do you load a CSV file into a Pandas DataFrame?

To load a CSV file into a Pandas DataFrame, the read_csv() function is used. This function reads the data from a CSV file and converts it into a structured DataFrame, which allows easy manipulation and analysis. The user needs to first import the Pandas library and then specify the file path of the CSV file. For example, using pd.read_csv("file_name.csv") loads the file into a DataFrame. This method is widely used because it is simple, efficient, and supports various parameters such as handling delimiters, headers, and missing values during the loading process.

4. What is the difference between head() and tail() functions in Pandas?

In Pandas, the head() and tail() functions are used to quickly view a subset of data from a DataFrame. The head() function returns the first few rows of the dataset, typically the first five rows by default, which helps in understanding the structure and initial data entries. On the other hand, the tail() function returns the last few rows of the dataset, also five rows by default, which is useful for inspecting the end of the data. Both functions are helpful for quickly checking data without displaying the entire dataset, especially when working with large datasets.

5. How can you identify missing values in a dataset?

Missing values in a dataset can be identified using various functions provided by the Pandas library. The most commonly used method is the isnull() or isna() function, which returns a boolean DataFrame indicating the presence of missing values. To count the total number of missing values in each column, the sum() function can be used along with isnull(), such as df.isnull().sum(). Additionally, functions like info() can provide a summary of non-null values in each column, helping to detect missing data. Identifying missing values is an important step in data cleaning, as it allows the user to decide how to handle them, such as removing or imputing the missing entries.

### **6. Remove Duplicate rows from a dataset**

```python
df = df.drop_duplicates()
df.shape
```

### **7. Handel missing values using mean or median**

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```
