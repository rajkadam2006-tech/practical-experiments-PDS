**THEORY :**

1. Difference between Binomial, Poisson, and Normal Distributions

The Binomial, Poisson, and Normal distributions are three important probability distributions used in statistics, but they differ in their nature and applications. The Binomial distribution is a discrete distribution that is used when there are a fixed number of independent trials, each having two possible outcomes such as success or failure. It is defined by parameters like the number of trials (n) and probability of success (p). The Poisson distribution is also discrete, but it is used to model the number of times an event occurs within a fixed interval of time or space, especially when events occur randomly and independently at a constant average rate. It is defined by a single parameter λ (lambda), which represents the average rate of occurrence. On the other hand, the Normal distribution is a continuous distribution that represents real-valued data and is characterized by its bell-shaped, symmetric curve. It is defined by mean (μ) and standard deviation (σ). While Binomial and Poisson deal with countable outcomes, Normal distribution is used for continuous measurements.

2. Real-world significance of Poisson Distribution

The Poisson distribution has significant real-world applications as it is used to model the occurrence of rare and random events over a fixed interval. It is particularly useful in situations where we want to count how many times an event happens within a given period of time or space. For example, in cricket, it can be used to model the number of wickets taken by a bowler in a match. Similarly, it is widely used in fields such as telecommunications (number of calls received per minute), traffic engineering (number of accidents at a location), and web analytics (number of website visits per second). Its importance lies in its simplicity and ability to provide accurate predictions when dealing with independent events that occur randomly but at a constant average rate.

3. Central Limit Theorem and its relation to Normal Distribution

The Central Limit Theorem (CLT) is a fundamental concept in statistics which states that when we take a large number of independent random samples from any distribution, the distribution of the sample means tends to become approximately Normal, regardless of the original distribution. This means that even if the original data follows a Binomial or Poisson distribution, the average of sufficiently large samples will follow a Normal distribution. The CLT explains why the Normal distribution appears so frequently in real-world data and statistical analysis. It allows statisticians to make inferences about population parameters using sample data and forms the basis for many statistical methods such as hypothesis testing and confidence intervals.

4. Why Normal Distribution is widely used in machine learning and statistics

The Normal distribution is widely used in statistics and machine learning because of its mathematical properties and its natural occurrence in real-world data. Many natural phenomena such as human heights, exam scores, and measurement errors tend to follow a Normal distribution. Due to the Central Limit Theorem, even when data does not initially follow a Normal distribution, the averages of samples tend to become normally distributed. This makes it extremely useful for analysis and modeling. Additionally, the Normal distribution is symmetric and easy to work with mathematically, which simplifies calculations in algorithms. In machine learning, many models assume normality in data or errors, making it a foundational concept in predictive modeling and statistical inference.

5. Skewness in distributions and which distributions exhibit it

Skewness refers to the measure of asymmetry in a probability distribution. A distribution is said to be positively skewed if the tail on the right side is longer, and negatively skewed if the tail on the left side is longer. If a distribution is perfectly symmetric, it has zero skewness. Among the distributions discussed, the Binomial distribution can be symmetric or skewed depending on the value of probability (p); it is symmetric when p = 0.5 and skewed otherwise. The Poisson distribution is generally positively skewed, especially when the average rate (λ) is small, meaning most values are clustered on the left with a long tail on the right. In contrast, the Normal distribution is perfectly symmetric and has no skewness, which is why it is often used as a standard reference distribution in statistics.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# -------------------- BINOMIAL (Boundaries in an innings) --------------------
# A batsman faces 10 balls, probability of hitting boundary = 0.5
balls = 10
p_boundary = 0.5

x_bin = np.arange(0, balls + 1)
bin_pmf = binom.pmf(x_bin, balls, p_boundary)
bin_samples = binom.rvs(balls, p_boundary, size=1000)

# -------------------- POISSON (Wickets per match) --------------------
# Average wickets per match = 2.5
lambda_wickets = 2.5

x_poi = np.arange(0, 10)
poi_pmf = poisson.pmf(x_poi, lambda_wickets)
poi_samples = poisson.rvs(lambda_wickets, size=1000)

# -------------------- NORMAL (Runs scored by player) --------------------
# Average runs = 35, std deviation = 15
mean_runs = 35
std_runs = 15

x_norm = np.linspace(0, 100, 1000)
norm_pdf = norm.pdf(x_norm, mean_runs, std_runs)
norm_samples = norm.rvs(mean_runs, std_runs, size=1000)

# -------------------- PLOTTING --------------------
plt.figure(figsize=(15, 10))

# Binomial
plt.subplot(3, 2, 1)
plt.bar(x_bin, bin_pmf)
plt.title("Binomial: Boundaries per 10 balls")

plt.subplot(3, 2, 2)
plt.hist(bin_samples, bins=range(12), density=True)
plt.title("Binomial Simulation")

# Poisson
plt.subplot(3, 2, 3)
plt.bar(x_poi, poi_pmf)
plt.title("Poisson: Wickets per match")

plt.subplot(3, 2, 4)
plt.hist(poi_samples, bins=range(11), density=True)
plt.title("Poisson Simulation")

# Normal
plt.subplot(3, 2, 5)
plt.plot(x_norm, norm_pdf)
plt.title("Normal: Runs Distribution")

plt.subplot(3, 2, 6)
plt.hist(norm_samples, bins=20, density=True)
plt.title("Normal Simulation")

plt.tight_layout()
plt.show()

# -------------------- STATISTICS --------------------
print("Binomial (Boundaries): Mean =", np.mean(bin_samples), "Variance =", np.var(bin_samples))
print("Poisson (Wickets): Mean =", np.mean(poi_samples), "Variance =", np.var(poi_samples))
print("Normal (Runs): Mean =", np.mean(norm_samples), "Std Dev =", np.std(norm_samples))
```
