**Theory :**
1. Explain the role of NumPy in the Python scientific computing ecosystem.

  NumPy (Numerical Python) is a fundamental library in the Python scientific computing ecosystem that provides support for large, multi-dimensional arrays and matrices along with a wide range of mathematical functions to operate on them. It acts as the backbone for many other libraries such as Pandas, Matplotlib, and Scikit-learn. NumPy enables efficient storage and processing of numerical data, making it highly suitable for tasks like data analysis, machine learning, and scientific simulations. Its powerful array structure allows users to perform complex computations in a simple and efficient manner, which is why it is widely used in research and industry applications.

2. Why is NumPy considered faster than pure Python computations?

NumPy is considered faster than pure Python computations because it uses optimized C and C++ code internally, which allows it to perform operations more efficiently. Unlike Python lists, NumPy arrays store data in contiguous memory locations, reducing overhead and improving performance. Additionally, NumPy uses vectorization, which means operations are applied to entire arrays at once instead of using loops. This eliminates the need for explicit iteration in Python, significantly speeding up computations. As a result, NumPy is highly efficient for handling large datasets and performing complex mathematical operations.

3. What types of mathematical operations can be performed using NumPy?

NumPy supports a wide variety of mathematical operations, including basic arithmetic operations such as addition, subtraction, multiplication, and division. It also provides advanced mathematical functions like trigonometric operations (sin, cos), logarithmic functions, exponentiation, and statistical operations such as mean, median, and standard deviation. NumPy also supports linear algebra operations including matrix multiplication, determinant calculation, and eigenvalues. These capabilities make NumPy a powerful tool for performing both simple and complex numerical computations efficiently.

4. Explain indexing and slicing in NumPy arrays

Indexing and slicing in NumPy arrays allow users to access and manipulate specific elements or subsets of data efficiently. Indexing refers to accessing a single element using its position, while slicing refers to extracting a portion of the array using a range of indices. NumPy supports both one-dimensional and multi-dimensional indexing, making it easy to work with complex data structures. Slicing is especially useful when working with large datasets, as it allows users to select only the required portion without copying the entire array.

5. How does NumPy integrate with libraries like Pandas, Matplotlib, and Scikit-learn?

NumPy integrates seamlessly with other popular Python libraries such as Pandas, Matplotlib, and Scikit-learn, making it a core component of the data science ecosystem. Pandas uses NumPy arrays as its underlying data structure, enabling efficient data manipulation and analysis. Matplotlib relies on NumPy arrays for plotting graphs and visualizing numerical data. Scikit-learn, a machine learning library, uses NumPy arrays to handle datasets and perform computations for training models. This integration allows smooth data flow between libraries and enhances performance and flexibility in data science workflows.

NmPy vs Python Speed Comparsion

This cell compares the execution speed of pure Python and NumPy. First, a Python list is created and each element is multiplied by 2 using a loop-like operation (list comprehension). Then, a NumPy array is created and multiplied by 2 using vectorized operations. The time module is used to measure the execution time for both approaches. This demonstrates that NumPy is faster because it performs operations at a lower level using optimized code and avoids explicit loops.





```python
import time

# Python list computation
start = time.time()
lst = list(range(1000000))
lst = [x * 2 for x in lst]
end = time.time()
print("Python Time:", end - start)

# NumPy computation
start = time.time()
arr = np.arange(1000000)
arr = arr * 2
end = time.time()
print("NumPy Time:", end - start)
```

Basic Mathematical Operations

In this cell, a NumPy array is created and basic arithmetic operations such as addition, subtraction, multiplication, and division are performed on it. These operations are applied element-wise, meaning each element in the array is processed individually. This shows how NumPy simplifies mathematical computations and allows operations to be performed on entire datasets efficiently.

```python
arr = np.array([1, 2, 3, 4, 5])

print("Original Array:", arr)
print("Addition:", arr + 2)
print("Subtraction:", arr - 1)
print("Multiplication:", arr * 3)
print("Division:", arr / 2)
```

Advanced Mathematical Functions

This cell demonstrates advanced mathematical and statistical functions provided by NumPy. Functions like sqrt() calculate square roots, power() raises elements to a specified power, and statistical functions like mean(), median(), and std() compute important measures of central tendency and dispersion. These built-in functions make NumPy highly useful for data analysis and scientific computing.

```python
print("Square Root:", np.sqrt(arr))
print("Power:", np.power(arr, 2))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

Indexing and Slicing (1D Array)

This cell explains how to access elements in a one-dimensional NumPy array using indexing and slicing. Indexing is used to retrieve a specific element by its position, while slicing is used to extract a subset of elements using a range. For example, arr[1:4] returns elements from index 1 to 3. This helps in efficiently working with subsets of data.

```python
arr = np.array([10, 20, 30, 40, 50])

print("First Element:", arr[0])
print("Last Element:", arr[-1])
print("Slice (1 to 3):", arr[1:4])
```

Indexing in 2D Array

In this cell, a two-dimensional NumPy array (matrix) is created, and different indexing techniques are demonstrated. Elements can be accessed using row and column indices, such as arr2[1, 2]. Additionally, entire rows or columns can be extracted using slicing. This is useful when working with structured data like tables or matrices.

```python
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print("Element at row 1, col 2:", arr2[1, 2])
print("First Row:", arr2[0])
print("Second Column:", arr2[:, 1])
```

NumPy with Pandas

This cell shows how NumPy integrates with the Pandas library. A NumPy array is created and then converted into a Pandas DataFrame. Pandas uses NumPy arrays internally, making data manipulation efficient. This integration allows users to easily move data between NumPy and Pandas for analysis and processing.

```python
import pandas as pd

data = np.array([1, 2, 3, 4, 5])
df = pd.DataFrame(data, columns=["Values"])

print(df)
```

NumPy with Matplotlib

In this cell, a NumPy array is used to create a simple line plot using Matplotlib. The array values are plotted on a graph, and labels are added for better visualization. This demonstrates how NumPy works with Matplotlib to visualize numerical data, which is important for data analysis and interpretation.

```python
import matplotlib.pyplot as plt

data = np.array([1, 2, 3, 4, 5])

plt.plot(data)
plt.title("NumPy Data Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

NumPy with Scikit-learn

This cell demonstrates how NumPy is used with Scikit-learn for machine learning tasks. A dataset is created using NumPy arrays, and a Linear Regression model is trained using this data. The model then predicts a new value based on input. This shows how NumPy provides the data structure required for machine learning algorithms in Scikit-learn.

```python
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
prediction = model.predict([[6]])
print("Prediction for 6:", prediction)
```
