**THEORY :**

1. What is correlation? How is it different from regression?

Correlation is a statistical measure that describes the strength and direction of the relationship between two variables. It indicates whether variables move together (positive correlation), move in opposite directions (negative correlation), or have no relationship at all. However, correlation does not imply causation and does not provide a predictive equation. In contrast, regression is a statistical technique used to model the relationship between a dependent variable and one or more independent variables. It not only shows the relationship but also provides a mathematical equation that can be used for prediction. In simple terms, correlation measures association, while regression focuses on prediction and estimation.

2. What is simple linear regression? Explain with suitable example.

Simple linear regression is a statistical method used to model the relationship between two variables by fitting a straight line to the observed data. One variable is considered independent (predictor), and the other is dependent (response). The relationship is expressed in the form of an equation:
Y = a + bX, where a is the intercept and b is the slope of the line.

For example, in cricket, the number of runs scored by a player (dependent variable) can be predicted based on the number of balls faced (independent variable). If a player scores more runs as the number of balls increases, a linear relationship can be established, and regression helps predict future performance based on past data.

3. How do you calculate Pearson correlation in Python?

Pearson correlation coefficient is used to measure the linear relationship between two variables. In Python, it can be calculated using libraries such as NumPy or Pandas. For example, using NumPy, the function np.corrcoef(x, y) returns the correlation coefficient between two datasets. Similarly, in Pandas, the .corr() method can be used on a DataFrame to compute correlation. The value of Pearson’s coefficient ranges from -1 to +1, where +1 indicates perfect positive correlation, -1 indicates perfect negative correlation, and 0 indicates no linear relationship.

4. What is multicollinearity? Does it affect simple regression?

Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. This makes it difficult to determine the individual effect of each independent variable on the dependent variable. Multicollinearity mainly affects multiple regression models where more than one predictor is used. In simple linear regression, there is only one independent variable, so multicollinearity does not occur. However, in more complex models, it can lead to unstable coefficients and unreliable predictions.

5. A dataset shows strong correlation but poor prediction accuracy. Why?

A dataset may show strong correlation but still have poor prediction accuracy due to several reasons. First, correlation only measures the strength of a linear relationship and does not guarantee that the relationship is useful for prediction. There may be outliers or noise in the data that affect the regression model. Second, the relationship might not be truly linear, even if correlation appears strong. Third, overfitting or underfitting of the model can reduce prediction accuracy. Additionally, missing variables or external factors not included in the model can influence the outcome, leading to inaccurate predictions despite high correlation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
balls_faced = np.array([8, 12, 18, 22, 28, 33, 38, 42, 48, 55]).reshape(-1, 1)
runs_scored = np.array([10, 15, 20, 27, 30, 36, 42, 47, 53, 61])

print("Balls:", balls_faced.flatten())
print("Runs:", runs_scored)
```

```python
correlation, _ = pearsonr(balls_faced.flatten(), runs_scored)
print("Pearson Correlation Coefficient:", round(correlation, 3))
```

```python
model = LinearRegression()
model.fit(balls_faced, runs_scored)

print("Intercept (a):", round(model.intercept_, 2))
print("Slope (b):", round(model.coef_[0], 2))
```

```python
predicted_runs = model.predict(balls_faced)

print("Predicted Runs:", predicted_runs)
```

```python
plt.figure(figsize=(8,6))

# Actual data points
plt.scatter(balls_faced, runs_scored)
plt.xlabel("Balls Faced")
plt.ylabel("Runs Scored")
plt.title("Cricket Data: Correlation & Regression")

# Regression line
plt.plot(balls_faced, predicted_runs)

plt.grid()
plt.show()
```

```python
r2 = r2_score(runs_scored, predicted_runs)
print("R² Score:", round(r2, 3))
```
